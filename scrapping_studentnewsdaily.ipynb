```python
url = "https://www.studentnewsdaily.com/archive/daily-news-article/"
```

```python
#call api server using requests module
result = requests.get(url)
#print(result)
#print(result.content)
soup = BeautifulSoup(result.content, 'lxml')
```

```python
#print(soup)
content = soup.find(class_= 'col-md-10 col-lg-10 col-xl-10')
#print(len(content))
print(content)
```

```python
heading_list = soup.findAll('h4') 
#print(heading_list)
```

```python
series = [element.get_text() for element in heading_list]
print(series)
```
