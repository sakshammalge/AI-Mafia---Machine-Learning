```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
X, Y = make_blobs(n_samples = 1000, n_features = 2, centers = 5, random_state = 6)
```

```python
plt.scatter(X[:,0],X[:,1],c = Y)
plt.show()
```

```python
k = 5
colours = ["green", "blue", "red", "orange", "gray"]
clusters = {}
for each_point in range(k):
    center = 10 * ((2 * np.random.random((X.shape[1],))) - 1)
    point = []
    cluster = {
        'center' : center,
        'colour' : colours[each_point],
        'point' : point
    }
    clusters[each_point] = cluster
```

```python
##funtions
```

```python
def dist(pt1, pt2):
    return np.sqrt(sum((pt1 - pt2)**2))
```

```python
def assign_points(clusters):
    for i in range(X.shape[0]):
        Pt = X[i]
        distance = []
        for j in range(k):
            d = dist(Pt, clusters[j]['center'])
            distance.append(d)
            
        current_cluster = np.argmin(distance)
        clusters[current_cluster]['point'].append(Pt)
```

```python
def Update_center(clusters):
    for i in range(k):
        pts = np.array(clusters[i]['point'])
        if pts.shape[0] > 0:
            new_mean = pts.mean(axis = 0)
            clusters[i]['center'] = new_mean
            clusters[i]['point'] = []
```

```python
def plot_cluster(clusters):
    for i in range(k):
        pts = np.array(clusters[i]['point'])
        if pts.shape[0] > 0:
            plt.scatter(pts[:,0], pts[:,1], color = clusters[i]['colour'])
        plt.scatter(clusters[i]['center'][0], clusters[i]['center'][1], color = "black", marker = "*")
```

```python
for t in range(50):
    assign_points(clusters)
    plot_cluster(clusters)
    Update_center(clusters)
```
