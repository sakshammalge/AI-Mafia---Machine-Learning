```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("garden.jpeg")
#print(img)
plt.imshow(img)
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
print(img.shape)
```

```python
pixels = img.reshape((1800*2880, 3))
print(pixels.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
k = 5 #Dominant Colors
km = KMeans(n_clusters = k)
km.fit(pixels)
```

```python
centers = km.cluster_centers_
centers = np.array(centers, dtype = "uint8")
print(centers)
```

```python
i=1
color = []
for each_col in centers:
    plt.subplot(1, k, i)
    plt.axis("off")
    i+=1
    color.append(each_col)
    
    a = np.zeros((100, 100, 3), dtype = "uint8")
    a[:,:,:] = each_col
    plt.imshow(a)

plt.show()
```

```python
#new Image
```

```python
new_img = np.zeros((1800*2880, 3), dtype = "uint8")
print(new_img.shape)
```

```python
labels = km.labels_
print(labels.shape)
```

```python
for each_pixel in range(new_img.shape[0]):
    new_img[each_pixel] = color[labels[each_pixel]]
    
#print(new_img)
new_img = new_img.reshape((img.shape))
```

```python
plt.imshow(img)
plt.show()
plt.imshow(new_img)
plt.show()
```
