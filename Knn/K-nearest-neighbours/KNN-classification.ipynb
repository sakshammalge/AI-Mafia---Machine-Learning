```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
plt.style.use("seaborn")
```

```python
X = pd.read_csv("../Datasets/xdata.csv")
X = X.values
X = X[:,1:]
Y = pd.read_csv("../Datasets/ydata.csv")
Y = Y.values
Y = Y[:,1:]
Y = Y.reshape((-1,))
```

```python
def distance(x1, x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
# Test Time 
def knn(X,Y,queryPoint,k=5):
    
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = distance(queryPoint,X[i])
        vals.append((d,Y[i]))
        
    
    vals = sorted(vals)
    vals = vals[:k]
    
    vals = np.array(vals)
    
    new_vals = np.unique(vals[:,1],return_counts=True)
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    
    return pred
```

```python
ans = knn(X,Y,[1,1])
```
