```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_x = pd.read_csv('./data/linearX.csv')
df_y = pd.read_csv('./data/linearY.csv')
```

```python
dfx = df_x.values
dfy = df_y.values

x = dfx.reshape((-1,1))#reshape for a single numpy array
y = dfy.reshape((-1,1))
```

```python
#print(x)
```

```python
plt.scatter(x,y,color='orange')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
output = model.predict(x)
```

```python
bias = model.intercept_
coef = model.coef_

print(bias)
print(coef)
```

```python
model.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,output, color ="red")
plt.show()
```
